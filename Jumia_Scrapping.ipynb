### Integrating Beautiful Soup

```python
# pass response to beautiful soup
# 3 types of parsers available - lxml, html.parser and html5lib
# lxml is most used as it is fast and handles complex even broken html

soup = BeautifulSoup(response.text, 'html.parser')
```

# find() and find_all() functions
find function is used to search for the first occurence of a specified tag within a parsed HTML Document
It returns the first matching tag found in the document structure

#syntax
find(name, attrs)
find_all(name, attrs)

```python
soup.find('h3',class_ = 'name')
```

```python
#Using dictionary to pass multiple attribute
soup.find('h3', {'class': 'name'})
```

```python
laptops = soup.find_all('h3', {'class':'name'})
```

```python
for i in laptops:
    print(i.text)
```

```python
## getting the price

prices = soup.find_all('div', {'class':'prc'})
```

```python
for i in prices:
    print(i.text)
```

```python
## getting the old price

old_price = soup.find_all('div', {'class':'old'})
```

```python
for i in old_price:
    print(i.text)
```

### Pagination in webscrapping

```python
# To loop through all the pages

all_laptops = []
```

```python
for i in range(10):
    response = requests.get(url, params = {'page': i+1})
    soup = BeautifulSoup(response.text, 'html.parser')
    page_laptops = soup.find_all('h3', {'class':'name'})
    all_laptops.append(soup)

```

### Beautiful Soup and Pandas

```python
r = requests.get(url)
```

```python
r.status_code
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
## all products 
laptops = []
products = soup.find_all('h3', {'class': 'name'})

```

```python
for product in products:
    laptops.append(product.text)
```

```python
laptops
```

```python
prices = []

all_prices = soup.find_all('div', {'class':'prc'})

for price in all_prices:
    prices.append(price.text)
```

```python
prices
```

```python
## old prices 

old_prices = []

all_old = soup.find_all('div', {'class':'old'})

for old in all_old:
    old_prices.append(old.text)
```

```python
old_prices
```
